```python
def get_recommend( input_list, recommend_table):
    L1 = len(input_list)
    L2 = 12
    T = L1+L2
    s = 0
    LL = len(recommend_table)
    idx_start = np.random.randint(LL)
    rec = recommend_table[idx_start]
    for i in recommend_table:
        s_tmp = T - len(set(i+input_list))
        if s_tmp>s:
            rec = i
            s = s_tmp
    return rec, s
```
